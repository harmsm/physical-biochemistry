```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

import scipy.special
```

```python
fig, ax = plt.subplots(3,4,figsize=(15,9),sharex="col")

N = [10,30,100,300]
for i in range(4):
    p = 1/6.
    max_k = N[i]
    
    k_range = np.arange(max_k+1,dtype=np.int)

    out1 = []
    out2 = []
    out3 = []
    for k in k_range:
        out1.append(scipy.special.comb(max_k,k))
        out2.append((p**k)*((1-p)**(max_k-k)))
        out3.append(out1[-1]*out2[-1])

    ax[0,i].plot(k_range/max_k,out1)
    ax[1,i].plot(k_range/max_k,out2)
    ax[2,i].plot(k_range/max_k,out3)
    ax[0,i].set_title("N = {}".format(N[i]))
    #
    ax[1,i].set_yscale("log")
    
    ax[0,0].set_ylabel("num. cocmbinations")
    ax[1,0].set_ylabel("$p^{k}(1-p)^{n-k}$")
    ax[2,0].set_ylabel("probability")
    ax[2,i].set_xlabel('fraction six')

plt.savefig("adding-more-N.png")
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy.special import comb
```

```python
def p_bound(n,k,p):
    
    return comb(n,k)*((p)**k)*((1-p)**(n-k))

step = 0.01
probs = np.arange(0,1+step,step)

zero_bound = []
one_bound = []
two_bound = []
three_bound = []
four_bound = []
for p in probs:
    
    zero_bound.append(p_bound(4,0,p))
    one_bound.append(p_bound(4,1,p))
    two_bound.append(p_bound(4,2,p))
    three_bound.append(p_bound(4,3,p))
    four_bound.append(p_bound(4,4,p))

fig, ax = plt.subplots(figsize=(5,5),frameon=False)
    
ax.plot(probs,zero_bound,linewidth=3,label="no O2")
ax.plot(probs,one_bound,linewidth=3,label="1 O2")
ax.plot(probs,two_bound,linewidth=3,label="2 O2")
ax.plot(probs,three_bound,linewidth=3,label="3 O2")
ax.plot(probs,four_bound,linewidth=3,label="4 O2")
ax.legend(loc="upper center",borderaxespad=2,frameon=False)

ax.set_xlabel("prob O2 binds at single site)",fontsize=14)
ax.set_ylabel("proportion species",fontsize=14)
ax.set_xlim(0,1.02)
ax.set_ylim(0,1.02)

for s in ax.get_xticklabels() + ax.get_yticklabels():
    s.set_fontsize(14)

ax.set_aspect('equal', 'box')
fig.savefig("naive-binding.png")
```

```python

O = 10**(np.arange(-10,1.01,0.01))

Ks = 0.01
K = 50


M = 1
Ms = Ks*(K*O)**0
M1 = Ks*(K*O)**1
M2 = Ks*(K*O)**2
M3 = Ks*(K*O)**3
M4 = Ks*(K*O)**4

Q = M + Ms + M1 + M2 + M3 + M4

P = K*O/(1 + K*O)


fig, ax = plt.subplots(figsize=(5,6),frameon=False)



ax.plot(P,(M + Ms)/Q,label="no O2",linewidth=3)
ax.plot(P,M1/Q,label="1 O2",linewidth=3)
ax.plot(P,M2/Q,label="2 O2",linewidth=3)
ax.plot(P,M3/Q,label="3 O2",linewidth=3)
ax.plot(P,M4/Q,label="4 O2",linewidth=3)
ax.legend(loc="center left",borderaxespad=2,frameon=False)
#ax.spines["top"].set_visible(False)
#ax.spines["right"].set_visible(False)
ax.set_xlabel("prob O2 binds at single site:\n(K*[O2])/(1 + K*[O2])",fontsize=14)
ax.set_ylabel("proportion species",fontsize=14)
ax.set_xlim(0,1.02)
ax.set_ylim(0,1.02)

for s in ax.get_xticklabels() + ax.get_yticklabels():
    s.set_fontsize(14)

ax.set_aspect('equal', 'box')

plt.savefig("hb-probs.png")

```
